```python
import numpy as np
import pandas as pd
```

```python
ir = np.genfromtxt('iris.data', delimiter=',', dtype = 'str')
```

```python
ir
```

```python
ir.shape
```

```python
uniqs = np.unique(ir[:,4:5])
ir2=ir
print(uniqs)
```

```python
np.argwhere(uniqs=='Iris-versicolor').tolist()[0][0]
```

```python
out = []
for val in uniqs:
    for s in ir2[ir2 == val]:
        group_id = np.argwhere(uniqs==s).tolist()[0][0]
        out.append(group_id )

print(out)
```
